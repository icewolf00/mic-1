```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Data Resource

```python
import pandas_datareader.data as web
from datetime import datetime
start = datetime(2014, 1, 1)
end = datetime.today()
baidu = web.DataReader('BIDU', 'iex', start, end)
alibaba = web.DataReader('BABA', 'iex', start, end)
facebook = web.DataReader('FB', 'iex', start, end)
apple = web.DataReader('AAPL', 'iex', start, end)
google = web.DataReader('GOOGL', 'iex', start, end)
microsoft = web.DataReader('MSFT', 'iex', start, end)
amazon = web.DataReader('AMZN', 'iex', start, end)
```

# Line Chart

```python
date_list = ['2014-01-02', '2015-01-02', '2016-01-04', '2017-01-03', '2018-01-02']
plt.figure(figsize=(20,10))
plt.plot(amazon['open'].loc[date_list], label='Amazon')
plt.plot(alibaba['open'].loc[date_list], label='Alibaba')
plt.plot(apple['open'].loc[date_list], label='Apple')
plt.plot(baidu['open'].loc[date_list], label='Baidu')
plt.plot(facebook['open'].loc[date_list], label='Facebook')
plt.plot(google['open'].loc[date_list], label='Google')
plt.plot(microsoft['open'].loc[date_list], label='Microsoft')
plt.legend()
plt.show()
```

# Pie Chart

```python
date = ['2018-01-02']
label_list = ['Alibaba', 'Amazon', 'Apple', 'Baidu', 'Facebook', 'Google', 'Microsoft']
pie_list = [alibaba['volume'].loc[date], amazon['volume'].loc[date], apple['volume'].loc[date], baidu['volume'].loc[date], facebook['volume'].loc[date], google['volume'].loc[date], microsoft['volume'].loc[date]]
plt.figure(figsize=(20,10))
plt.pie(pie_list, labels=label_list)
plt.show()
```

# Bar Chart

```python
bar_list = list()
date_list = ['2015-01-02', '2016-01-04', '2017-01-03', '2018-01-02']
plt.figure(figsize=(20,10))
def plt_bar(data, label):
    year_list = list()
    for i in date_list:
        year_list.append(int(data.loc[i]))
    plt.bar(date_list, year_list, label=label, width=0.3)
plt_bar(alibaba['volume'], 'Alibaba')
plt_bar(amazon['volume'], 'Amazon')
# plt_bar(apple['volume'])
plt_bar(baidu['volume'], 'Baidu')
# plt_bar(facebook['volume'])
plt_bar(google['volume'], 'Google')
# plt_bar(microsoft['volume'])
plt.legend()
plt.show()
```
